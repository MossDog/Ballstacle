```python
import math

def ease_out_elastic(x: float) -> float:
    c4 = (2 * math.pi) / 3

    if x == 0:
        return 0
    elif x == 1:
        return 1
    else:
        return math.pow(2, -10 * x) * math.sin((x * 10 - 0.75) * c4) + 1

```

```python
num = 0
for i in range(0, 100, 5):
    num += 1

    print(f"t = {i/100} -|- x = {ease_out_elastic(i/100)}")
```
